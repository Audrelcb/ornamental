```python
# Load the CSV file into a DataFrame
df3 = pd.read_csv(r'C:\\Users\Lecab\largefiles\cdhrc_dbgi\pos\merge_met_annot_sirius\merged_met_annot_sirius_canopus_GNPS.tsv', sep='\t')
# Initialize the 'score' column with default values of 1
df3['score'] = 1

# Iterate over the rows of the DataFrame
for index, row in df3.iterrows():
    inchikey_sirius_value = str(row['InChIkey2D_sirius']) if pd.notnull(row['InChIkey2D_sirius']) else ""
    inchikey_metAnnot_value = str(row['short_inchikey_metAnnot']) if pd.notnull(row['short_inchikey_metAnnot']) else ""
    inchikey_GNPS_value = str(row['InChIKey_GNPS']) if pd.notnull(row['InChIKey_GNPS']) else ""

    # Compare the values of the non-empty columns with a minimum of 14 characters
    if len(inchikey_sirius_value) >= 14 and len(inchikey_metAnnot_value) >= 14 and len(inchikey_GNPS_value) >= 14:
        if inchikey_sirius_value[:14] == inchikey_metAnnot_value[:14] == inchikey_GNPS_value[:14]:
            df3.at[index, 'score'] = 3
        elif inchikey_sirius_value[:14] == inchikey_metAnnot_value[:14] or inchikey_sirius_value[:14] == inchikey_GNPS_value[:14] or inchikey_metAnnot_value[:14] == inchikey_GNPS_value[:14]:
            df3.at[index, 'score'] = 2
    elif len(inchikey_sirius_value) >= 14 and len(inchikey_metAnnot_value) >= 14 and inchikey_sirius_value[:14] == inchikey_metAnnot_value[:14]:
        df3.at[index, 'score'] = 2
    elif len(inchikey_sirius_value) >= 14 and len(inchikey_GNPS_value) >= 14 and inchikey_sirius_value[:14] == inchikey_GNPS_value[:14]:
        df3.at[index, 'score'] = 2
    elif len(inchikey_metAnnot_value) >= 14 and len(inchikey_GNPS_value) >= 14 and inchikey_metAnnot_value[:14] == inchikey_GNPS_value[:14]:
        df3.at[index, 'score'] = 2

# Save the updated DataFrame to a new TSV file
df3.to_csv(r'C:\\Users\Lecab\largefiles\cdhrc_dbgi\pos\merge_met_annot_sirius\annotation_score.tsv', sep='\t', index=False)
```
